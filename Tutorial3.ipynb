```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for enhanced visualizations

df = pd.read_csv("titanic.csv")
display(df.head())
df.info()

print("\nüîç Checking missing values in each column:")
display(df.isnull().sum())

# Fill missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df["Cabin"] = df["Cabin"].fillna("")


print("\n Descriptive Statistics:")
display(df.describe())

print("\n Value Counts for Categorical Columns:")
print("Survived value counts:")
display(df["Survived"].value_counts())
print("\nSex value counts:")
display(df["Sex"].value_counts())
print("\nEmbarked value counts:")
display(df["Embarked"].value_counts())


print("\n Histograms of Numerical Columns:")
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

print("\n Bar charts of categorical columns:")
categorical_cols = ['Sex', 'Embarked']

for col in categorical_cols:
  freq_table = df[col].value_counts()
  plt.figure(figsize=(7, 5))
  freq_table.plot(kind='bar', color='lightcoral', edgecolor='black')
  plt.title( f"Bar Chart of {col}")
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.xticks(rotation=45)
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.show()

#find outliers and remove them
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_cleaned

df = remove_outliers_iqr(df, "Age")
df = remove_outliers_iqr(df, "Fare")


print("\n Bivariate Analysis:")

plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Scatter Plot: Age vs Fare (colored by Survival)")
plt.show()

plt.figure(figsize=(7,5))
sns.boxplot(x="Sex", y="Age", hue="Survived", data=df)
plt.title("Boxplot: Fare by Passenger Class")
plt.show()

print("\n Correlation matrix:")

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# üéØ Insight 1: Survival by Sex
print("\nüéØ Insight 1")
print("\n Strongest predictor of survival. Most females survived; most males did not.")
plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Sex")
plt.ylabel("Survival Rate")
plt.show()

# üéØ Insight 2: Survival by Pclass
print("\nüéØ Insight 2")
print("\n 1st class passengers had much better survival odds.")

plt.figure(figsize=(6,4))
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

# üéØ Insight 3: Survival by Age Group
df['Age_bin'] = pd.cut(df['Age'], bins=[0, 12, 18, 40, 60, 80], labels=['Child','Teen','Adult','Middle Aged','Senior'])
print("\n Younger passengers had better survival, especially children. ")
print("\nüéØ Insight 3")
print("\n Younger passengers had better survival, especially children. ")
plt.figure(figsize=(8,4))
sns.barplot(x="Age_bin", y="Survived", data=df)
plt.title("Survival Rate by Age Group")
plt.ylabel("Survival Rate")
plt.show()
3
```
